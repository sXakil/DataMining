```python
def pearson(x,y):
    correlation_matrix = np.corrcoef(x,y)
    return correlation_matrix[0,1]
```

```python
r=pearson(df.SepalLengthCm,df.SepalWidthCm)
```

```python
print(r)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
csimilarity = cosine_similarity([df.SepalLengthCm],[df.SepalWidthCm])
```
