```python
import pandas as pd
```

```python
df = pd.read_csv('../data/iris_versicolor.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.SepalLengthCm, df.SepalWidthCm, 'bo')
plt.xlabel('length')
plt.ylabel('width')
```

```python
import numpy as np
```

```python
cov = np.cov(df.SepalLengthCm, df.SepalWidthCm)
print(cov)
```

```python
def pearson(x, y):
    return np.corrcoef(x, y)[0,1]
```

```python
p = pearson(df.SepalLengthCm, df.SepalWidthCm)
print(p)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs  = cosine_similarity([df.SepalLengthCm], [df.SepalWidthCm])
print(cs)
```
